```python

# Linear graph 
import matplotlib.pyplot as plt
plt.title('Basic Linear Graph')
plt.plot([1,2,3],[1,2,3])
plt.show()
```

### Graph to plot distance and time

```python
distance = range(2,11,2)
time = range(1,6)
plt.xlabel("Time")
plt.ylabel("Distance")
plt.plot(time,distance,'bo')
plt.grid()
plt.title('Distance Time Graph')
plt.show()
```

```python
# Non-Linear Graph

# v = a t

t = list(range(1,11))
a = list(range(1,len(t)+1))
v = [a[i] * i for i in range(len(t))]
plt.plot(t,v,'bo')
plt.grid()
plt.title('Time VS Velocity')
plt.xlabel('Time')
plt.ylabel('Velocity')
plt.show()
```

```python
# Graphs for Numbers Vs Squares

n = list(range(10))
sn = [i**2 for i in n]

plt.plot(n,sn)
# plt.grid()
plt.title('Numbers Vs Squares')
plt.xlabel('Numbers')
plt.ylabel("Squares")
plt.show()
```

```python
# Sinsoidal Graph

# Voltage VS Time
import numpy as np

t = [i/10 for i in range(1,21)]
v = [1 + np.sin(2*np.pi*i) for i in t]
plt.plot(t,v,'ro')
plt.title('Voltage vs Time')
plt.grid()
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.savefig('Images/Voltage-Time.png')
plt.show()
```

```python
# Plot Income of Arkansas from 2005 to 2013

import pandas as pd
df = pd.read_csv('DataFiles\Income.csv')
years = df.columns[2:]
income = df.values[3,2:]
plt.title('Income of Arkansas')
plt.xlabel('Year')
plt.ylabel('Income')
plt.plot(years,income)
plt.show()
```

```python
# Income of all states from 2005 to 2013
incomeallstates = []
for i in df.columns[2:]:
    incomeallstates.append(df[i].mean())
plt.plot(years,incomeallstates)
plt.title('All States Income')
plt.xlabel('Year')
plt.ylabel('Time')
plt.show()
```
